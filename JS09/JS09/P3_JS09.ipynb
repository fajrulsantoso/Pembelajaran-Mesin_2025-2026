Langkah 1 - Load Data


```python
import numpy as np
import pandas as pd

# Langkah 1 - Load Data
df = pd.read_csv('spam.csv', encoding='latin-1')
df.head()

# Langkah 2a - Drop kolom yang tidak digunakan
df = df.drop(df.iloc[:, 2:], axis=1)

# Rename kolom (v1 → Labels, v2 → SMS)
df = df.rename(columns={'v1': 'Labels', 'v2': 'SMS'})

# Langkah 2b - Inspeksi Data
print(df['Labels'].value_counts())
print('\n')
print(df.info())
print('\n')
print(df.describe())

# Langkah 2c - Encoding Label
new_labels = {
    'spam': 1,
    'ham': 0
}
df['Labels'] = df['Labels'].map(new_labels)
df.head()

# Langkah 2d - Pisahkan fitur dengan label
X = df['SMS'].values
y = df['Labels'].values

# Langkah 3 - Ekstraksi Fitur
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=50
)

bow = CountVectorizer()

# Fit & transform data training
X_train = bow.fit_transform(X_train)

# Transform data testing
X_test = bow.transform(X_test)

# --- BARIS YANG SUDAH DIPERBAIKI ---
print(len(bow.get_feature_names_out()))
print(f'Dimensi data: {X_train.shape}')

# Langkah 4 - Training dan Evaluasi
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

mnb = MultinomialNB()
mnb.fit(X_train, y_train)

y_pred_train = mnb.predict(X_train)
acc_train = accuracy_score(y_train, y_pred_train)

y_pred_test = mnb.predict(X_test)
acc_test = accuracy_score(y_test, y_pred_test)

print(f'Hasil akurasi data train: {acc_train}')
print(f'Hasil akurasi data test: {acc_test}')
```
