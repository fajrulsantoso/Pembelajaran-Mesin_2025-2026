#Pisahkan variabel yang dapat dan tidak dapat digunakan



1.   Kolom id tidak berguna → dibuang
2.   Kolom diagnosis adalah target.
3.   Sisanya adalah fitur numerik yang dipakai.



```python
# Load dataset
path = '/content/drive/MyDrive/Marchine learning/wbc (1).csv'
df = pd.read_csv(path)

print("Kolom awal:", df.columns.tolist())

# Buang kolom id jika ada
if 'id' in df.columns:
    df = df.drop(columns=['id'])

print("\nKolom setelah id dibuang:", df.columns.tolist())
df.head()

```

#Lakukan proses encoding pada kolom diagnosis  

1. Diagnosis: M (Malignant) → 1, B (Benign) → 0



```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

print("Jumlah kelas setelah encoding:")
print(df['diagnosis'].value_counts())
```

#Lakukan proses standarisasi pada semua kolom numerik

1.   Standarisasi membuat data memiliki skala yang sama (mean=0, std=1)



```python
# Buang kolom tidak berguna (abaikan kalau tidak ada)
df = df.drop(columns=['id', 'Unnamed: 32'], errors='ignore')

# Pisahkan fitur dan target
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']

# Standardisasi fitur numerik
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Sebelum standarisasi (ringkasan 5 kolom pertama):")
display(X.describe().iloc[:, :5])

print("\nSetelah standarisasi (contoh 5 baris, 5 kolom):")
print(X_scaled[:5, :5])
```

#Lakukan proses stratified split data (80:20)

1.   Stratified split menjaga proporsi kelas tetap seimbang antara data latih dan data uji.



```python
# Split dengan stratifikasi
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print("Ukuran data latih:", X_train.shape)
print("Ukuran data uji:", X_test.shape)

print("\nProporsi label di train:")
print(y_train.value_counts(normalize=True))

print("\nProporsi label di test:")
print(y_test.value_counts(normalize=True))
```
